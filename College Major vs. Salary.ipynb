## Find College Major with Highest Starting Salaries



```python
max_starting_salary = clean_df["Starting Median Salary"].max()
print(max_starting_salary)
```

```python
clean_df[clean_df["Starting Median Salary"] == max_starting_salary]
```

We found the Major with the highest starting salary of $74300 is Physician Assistant.

## Find College Major with Highest Mid-Career Salaries

```python
max_mid_salary = clean_df["Mid-Career Median Salary"].max()
print(max_mid_salary)
```

```python
clean_df[clean_df["Mid-Career Median Salary"] == max_mid_salary]
```

Chemical engineering major has the highest mid-career salary with the median of $107000.

## Find College Major with Lowest Starting Salaries

```python
min_starting_salary = clean_df["Starting Median Salary"].min()
print(min_starting_salary)
```

```python
clean_df[clean_df["Starting Median Salary"] == min_starting_salary]
```

Spanish major has the lowest starting salary with the median of $34000.

```python
min_mid_salary = clean_df["Mid-Career Median Salary"].min()
print(min_mid_salary)
```

```python
clean_df[clean_df["Mid-Career Median Salary"] == min_mid_salary]
```

We have 2 majors with the lowest salary during their mid careers that are Education and Religion.

## Majors with the Most Potential vs Lowest Risk

A low-risk major is a degree where there is a small difference between the lowest and highest salaries. In other words, if the difference between the 10th percentile and the 90th percentile earnings of your major is small, then you can be more certain about your salary after you graduate.

Using that definition, we can easily find majors with th most potential as well as lowest risk.

```python
spread_col = clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']
clean_df.insert(1, 'Spread', spread_col)
clean_df.head()
```

```python
clean_df
```

### Majors with lowest risk

```python
low_risk = clean_df.sort_values('Spread')
low_risk[['Undergraduate Major', 'Spread']].head()
```

```python
most_potential = clean_df.sort_values('Spread', ascending=False)
most_potential[['Undergraduate Major', 'Spread']].head()
```

## Grouping data

which category of degrees has the highest average salary? Is it STEM, Business or HASS (Humanities, Arts, and Social Science)?

```python
clean_df.groupby("Group").count()
```

```python
pd.options.display.float_format = '{:,.2f}'.format 
clean_df.groupby("Group").mean()
```

We can see taht Business Majors have the highest potential while HASS group is low risk. STEM group has the best starting salary as well as mid-career salary. If you can, just tell your kids to follow STEM major.
